# データ処理2

```python
#移動平均 & shift
df['Total'] = df['UnitPrice'] * df['Quantity']
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
sales_1d = df.resample('D',on = 'InvoiceDate').sum()
sales_1d= sales_1d.drop(columns='CustomerID')
```

```python
#rolling
sales_1d['3d_moving_avg'] = sales_1d['Total'].rolling(3).mean()
sales_1d
```

```python
#rolling center
sales_1d['3d_moving_avg_ctr'] = sales_1d['Total'].rolling(3, center = True).mean()
sales_1d
```

```python
#shift
sales_1d['3d_moving_avg_shift1'] = sales_1d['Total'].rolling(3).mean().shift(1)
sales_1d
```

```python
#diff (A - B)
sales_1d['diff'] = sales_1d['Total'].diff()
sales_1d
```

```python
#pct_change((A - B) / B)
sales_1d['pct_change'] = sales_1d['Total'].pct_change()
sales_1d
```
